```python
import time 
def timeit(f):
    def newf(*args, **kwargs):
        start = time.time()
        retval = f(*args, **kwargs)
        end = time.time()
        return retval, round((end - start), 3)
    return newf
```

```python
"""
This is from Section 1.2 of Classic Computer Science Problems in Python,
by David Kopec.
"""

@timeit
def compress(dna:str, start: int, stop: int) -> int:
    bits = 0   # let m be the number of bits in bits

    d = {'a': 0b00, 'c': 0b01, 'g': 0b10, 't': 0b11}
    for i in range(start, stop):        # n = stop - start
        bits = (bits << 2) | d[dna[i]]  # m = 2n

    return bits




```
